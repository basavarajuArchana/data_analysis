# Basic plotting with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'b+')  # indicate the points with + symbol
plt.show()
```

```python
plt.plot(x,y,'bo',x,y,'k-') # points will be shown in 'o' and line color is black
plt.show()
```

```python
# axis labelling and display title

x=np.array([120,140,150,170,131])
y=np.array([66,82,44,67,94])
plt.plot(x,y,'ko')
plt.xlabel(('Heights (cm)'))
plt.ylabel('weights(kg)')
plt.title(" Heights and weights")
plt.show()
```

# Box plot

 1.Very useful to detect outliers.
 
 2.Shows 25th, 50th and 75th percentiles.
 
 3. Used for both Univariate analysis and bi variate analysis 

Percentile basically says where a point lies in a sorted list.

Median-> 50th percentile
Quantiles-> 25th, 50th, 75th, 100th percentile


75th percentile-> 75 percent of all the observations are less than 75th percentile.

Eg: CAT Rank
